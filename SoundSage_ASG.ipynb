```python
import matplotlib.pyplot as plt

def plot_smoothed_trends_grid(df, numeric_columns, group_col, rolling_window=5):
    """
    Plots smoothed trends for numerical columns in a grid layout.

    Parameters:
    df (pd.DataFrame): The dataframe containing the data.
    numeric_columns (list): List of numerical columns to plot.
    group_col (str): Column to group by (e.g., 'year').
    rolling_window (int): Window size for smoothing (default is 5).
    """
    # Calculate number of rows and columns for the grid
    n_cols = 3  # Number of columns in the grid
    n_rows = (len(numeric_columns) + n_cols - 1) // n_cols  # Calculate required number of rows

    # Create subplots
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows*5))

    # Flatten axes array to make indexing easier
    axes = axes.flatten()

    for idx, col in enumerate(numeric_columns):
        if col in df.columns:
            # Calculate mean per group and apply rolling mean
            smoothed = df.groupby(group_col)[col].mean().rolling(window=rolling_window, center=True).mean()

            # Plot
            axes[idx].plot(smoothed)
            axes[idx].set_title(f'Smoothed Average {col.capitalize()} Over {group_col.capitalize()}s')
            axes[idx].set_xlabel(group_col.capitalize())
            axes[idx].set_ylabel(col.capitalize())
            axes[idx].grid()

    # Hide any empty subplots
    for i in range(len(numeric_columns), len(axes)):
        axes[i].axis('off')

    plt.tight_layout()
    plt.show()

# Example usage
numeric_columns = ['valence', 'danceability', 'energy', 'tempo', 'duration_minutes']
plot_smoothed_trends_grid(df, numeric_columns, group_col='year', rolling_window=5)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Load your dataset
# Assuming `df` is your dataframe with numeric columns
numeric_columns = ['valence', 'danceability', 'energy', 'tempo', 'duration_ms']
data = df[numeric_columns]

# Normalize the data (Standardize to mean 0 and variance 1)
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Run DBSCAN
epsilon = 0.5  # Neighbor search radius
min_samples = 15  # Minimum points for a cluster
dbscan = DBSCAN(eps=epsilon, min_samples=min_samples)
clusters = dbscan.fit_predict(data_scaled)

# Add cluster labels to the dataframe
df['cluster'] = clusters

# Visualize the clusters (2D scatter plot)
plt.figure(figsize=(10, 6))
scatter = plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=clusters, cmap='viridis', alpha=0.6)
plt.colorbar(scatter, label='Cluster Label')
plt.title('DBSCAN Clustering Results')
plt.xlabel(numeric_columns[0])
plt.ylabel(numeric_columns[1])
plt.grid(True)
plt.show()

# Print summary of clusters
num_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)
num_noise = list(clusters).count(-1)
print(f'Total clusters formed: {num_clusters}')
print(f'Total noise points: {num_noise}')
```

```python
import random
import pandas as pd
import numpy as np

# Generate synthetic users with preferences based on feature ranges
def generate_users(num_users, dataset, num_likes=10):
    # Compute the min and max values for each feature
    feature_ranges = {
        feature: (dataset[feature].min(), dataset[feature].max())
        for feature in ['danceability', 'energy', 'tempo','loudness','speechiness','acousticness','instrumentalness','liveness']
    }

    users = []
    for user_id in range(num_users):
        # Generate preferences within the feature's range
        preferences = {
            feature: random.uniform(feature_ranges[feature][0], feature_ranges[feature][1])
            for feature in feature_ranges
        }
        user = User(user_id, dataset, num_likes, preferences)
        users.append(user)
    return users

users = generate_users(10000, df, num_likes=random.randint(4,8))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def compute_user_profile(user, dataset, features):
    liked_tracks = user.get_liked_tracks()
    preferences = user.get_preferences()

    # Filter liked tracks and extract their feature values
    liked_features = dataset[dataset['id'].isin(liked_tracks)][features].values

    # Apply preferences as weights
    weights = np.array([preferences.get(f, 1.0) for f in features])
    weighted_features = liked_features * weights

    # Compute the average profile
    user_profile = np.mean(weighted_features, axis=0)

    return user_profile
```

```python
def recommend_tracks(user, dataset, features, top_n=5):
    user_profile = compute_user_profile(user, dataset, features)

    # Normalize the user profile for cosine similarity
    user_profile = user_profile.reshape(1, -1)

    # Compute similarities with all tracks in the dataset
    track_features = dataset[features].values
    similarities = cosine_similarity(user_profile, track_features).flatten()

    # Add similarity scores to the dataset
    dataset['similarity'] = similarities

    # Exclude already liked tracks
    liked_tracks = user.get_liked_tracks()
    recommendations = dataset[~dataset['id'].isin(liked_tracks)].sort_values(by='similarity', ascending=False)

    return recommendations.head(top_n)
```

```python
features = ['danceability', 'energy', 'tempo', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness']
# Test on the first user
test_user = users[0]
recommended_tracks = recommend_tracks(test_user, df, features, top_n=5)
print(recommended_tracks[['id', 'name', 'similarity']])
```
